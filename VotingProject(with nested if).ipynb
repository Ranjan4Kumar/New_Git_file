```python
# we need to assign 2 candidate whom we are going to choose may be there are more than 2 candidates we can choose accordingly
candidate1 = input('Enter 1st candidate name: ' )
candidate2 = input('Enter 2nd candidate name: ' )
 # we need to create a variable and assign 0 to it which count the no. of vote for both the candidates
cndidate1 = 0
cndidate2 = 0
 # we will give some id to the the voters and we stored it inside one variable name voter_id
voter_id = [101,102,103]
voter_id_length = len(voter_id)
print('Total no. of voters will apear in voting is :',voter_id_length)

# we will create an empty list which will store the voters who voted and we remove id's from voter_id and store in this list
voted = []

# we will define a while loop which always be true
while True:
  if voter_id == []:
    print("Voting is over")
    if cndidate1>cndidate2:
      print(f"{candidate1} won the election with total no. of votes {cndidate1}")
    elif candidate1<candidate2:
      print(f"{candidate2} won the election with total no. of votes {cndidate2}")
    elif candidate1 == candidate2:
      print('result tie')
    break

  else:
    voter = int(input('Enter your id :' ))
    if voter in voted:
      print('You already voted you are not allowed')
    else:
      if voter in voter_id:
        print(f"1.{candidate1}\n2.{candidate2}")
        choice = int(input('Enter your choice as 1 or 2'))
        if choice == 1:
          cndidate1 +=1
          print(f"you voted {candidate1}")
        elif choice == 2:
          cndidate2 += 1
          print(f"you voted{candidate2}")
        voter_id.remove(voter)
        voted.append(voter)
      else:
        print("voter_id is missmatched type again otherwise you are not allowed to vote")



```
